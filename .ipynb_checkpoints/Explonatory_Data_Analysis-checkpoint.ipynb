# Let's have a look into win ratio of teams that play at their own stadion vs teams that play away

```python
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

count = df['Result'].value_counts()
count.plot(kind='bar')
plt.show()
print('Results of matches by Home Team')
print(count/len(df['Result']))
```

As might have been expected- teams perform better when thay play at their own stadion. Let's see if we can see those differences in match statistics.

```python
names = df.columns[5:27]
for i in range(0,len(names),2):
    plt.figure()
    df.boxplot(column=[names[i],names[i+1]])
```

From the last plot we can see that in 75% of matches Home teams scored 1 or more goals and it's also a median value. I think it's interesting and might be useful if we would like to predict number of scored goals in the match.

```python
df_num = df[names]
fig, ax = plt.subplots(nrows=11, ncols=1, figsize=(100,400))
j = 0
for i in range(0,len(names),2):
    ax[j].set_title(f'{df_num.columns[i][:-5]}',size = 80)
    ax[j].tick_params(axis='x', labelsize=80)
    ax[j].tick_params(axis='y', labelsize=80)
    ax[j].hist(df_num[f'{df_num.columns[i]}'], alpha = 0.5, label='Home')
    ax[j].hist(df_num.iloc[:,i+1], alpha = 0.5, label='Away')
    ax[j].legend(loc = 'upper right', fontsize = 100)
    
    j+=1
```

Let's check how many times Home team scored given number of goals

```python
goals_home_pct = (df.Goals_home.value_counts()/df.shape[0]).sort_index()
print('How many times Home team scored given number of goals')
print(goals_home_pct)
goals_home_cumpct = goals_home_pct.cumsum()
print("Cumulative pct")
print(goals_home_cumpct)
```

Let's see how it looks only in current season

```python
current_season = df.query('Season == "2022-2023"')
goals_home_pct = (current_season.Goals_home.value_counts()/current_season.shape[0]).sort_index()
print('How many times Home scored given number of goals (season 2022-2023)')
print(goals_home_pct)
goals_home_cumpct = goals_home_pct.cumsum()
print("Cumulative pct")
print(goals_home_cumpct)
```

Let's have a look into sum of goals scored by both teams in each match

```python
df_goals = df['Goals_home'] + df['Goals_away']
goals_pct = (df_goals.value_counts()/df.shape[0]).sort_index()
print(goals_pct)
goals_cumpct = goals_pct.cumsum()
print("Cumulative pct")
print(goals_cumpct)
```

As we can see, conlcusions from histograms are simillar like from boxplots. We can also notice that some of the variables are likely to be normally distributed.

# I think good way to build a model will be calculating some statistics from previous matches of each team to measure their performance and form for upcoming match. Let's add new variables to DataFrame which will be averages of statistics from previous 3 matches for both teams.

```python
df_1,df_2 = df.copy(), df.copy()
df_1['Team'] = df_1['Team_Home']
df_2['Team'] = df_2['Team_Away']
df_dup = pd.concat([df_1,df_2], axis=0)
print(df_dup)
```

```python
grouped_home = df.groupby(['Season','Team_Home'])
grouped_away = df.groupby(['Season','Team_Away'])
# group = grouped_home.get_group(('2012-2013','Lechia Gdańsk'))
```

```python
def roll_avg(group,cols, cols_new):
    group = group.sort_values(['Season','Round','Date'])
    roll_stats = group[cols].rolling(3,closed='left').mean()
    group[cols_new] = roll_stats
    # drop first 3 rounds of every season
#     group = group.dropna(subset=cols_new)
    return group
```

```python
cols_home = ['Ball_possession_home','Goal_sit_home','Shots_target_home','Shots_no_target_home','Corners_home','Offsides_home','GK_inter_home','Fauls_home','Yellow_home','Red_home','Goals_home']
roll_home = [f'r_{c}' for c in cols_home]
cols_away = ['Ball_possession_away','Goal_sit_away','Shots_target_away','Shots_no_target_away','Corners_away','Offsides_away','GK_inter_away','Fauls_away','Yellow_away','Red_away','Goals_away']
roll_away = [f'r_{c}' for c in cols_away]
```

```python
# rolling for home team
df_roll_1 = df_dup.groupby(['Season','Team']).apply(lambda x: roll_avg(x, cols_home, roll_home))
# rolling for away team
df_roll_2 = df_dup.groupby(['Season','Team']).apply(lambda x: roll_avg(x, cols_away, roll_away))

df_roll_home = df_roll_1[df_roll_1['Team'] == df_roll_1['Team_Home']]
df_roll_away = df_roll_2[df_roll_2['Team'] == df_roll_2['Team_Away']]
```

```python
df_roll_home.info()
df_roll_away.info()
```

```python
# prepare columns names that we need from df_roll_away
df_roll_home = df_roll_home.droplevel(['Season','Team'])
df_roll_away = df_roll_away.droplevel(['Season','Team'])
cols_df_away = ['Season','Round','Date','Team'] +  roll_away
print(cols_df_away)
```

```python
df_rolled = pd.merge(df_roll_home, df_roll_away[cols_df_away], left_on = ['Season','Round','Date','Team_Away'], right_on = ['Season','Round','Date','Team'])
```
