```python
# import libraries
import pandas as pd
import numpy as np
import glob
import os
import matplotlib.pyplot as plt
```

```python
# load csv files
path = os.getcwd()
all_files = glob.glob(os.path.join(path,"*.csv"))
df = pd.DataFrame(pd.concat((pd.read_csv(file) for file in all_files), ignore_index = True))
```

# Check how many matches do we have for each season

```python
plt.figure(figsize=(20,6))
plt.style.use('fivethirtyeight')
plt.hist(df['Season'], bins = 11, rwidth = 0.9)
plt.title('Number of matches in each season')
plt.xlabel('Season')
plt.ylabel('Number of matches')
plt.show()
```

It looks ok since for season 2012/13 I was scrapping only about 110 matches as only those matches had available statistics, for season 2020/21 I checked in the internet that there were less matches played and in the current season all matches hasn't been played yet.

# Check columns type and their names (and change them)

```python
print(df.columns)
df.info()
df.head()
```

```python
# save original dataset
df_orig = df
```

```python
# drop first column and chagne columns names
df = df_orig.iloc[:,1:]
df.columns = ['Season','Round','Date','Team_Home','Team_Away',
              "Ball_possession_home","Ball_possession_away",
              "Goal_sit_home","Goal_sit_away",
              "Shots_target_home","Shots_target_away",
              "Shots_no_target_home", "Shots_no_target_away",
              "Corners_home","Corners_away",
              "Offsides_home", "Offsides_away",
              "GK_inter_home", "GK_inter_away",
              "Fauls_home", "Fauls_away",
              "Yellow_home", "Yellow_away",
              "Red_home", "Red_away",
              "FK_home", "FK_away",
              "Goal_kicks_home", "Goal_kicks_away",
              "Goals_home", "Goals_away",
              "Blocked_shots_home", "Blocked_shots_away",
              "Auts_home", "Auts_away",
              "Attacks_home", "Attacks_away",
              "Dang_att_home","Dang_att_away",
              "Pass_home","Pass_away",
              "Pass_target_home", "Pass_target_away",
              "Blocks_home", "Blocks_away",
              "Acc_pass_home", "Acc_pass_away"]
print(df.columns)
print(df.head())
```

# Check missing values

```python
# check how many missing values do we have for each column
df_nulls = df.isnull().sum()/df.shape[0]

plt.figure(figsize=(20,20))
df_nulls.plot(kind='barh')
plt.title("% of missing values", size =40)
plt.yticks(fontsize=20)
plt.xlim([0,1.0])
plt.show()
```

Some variables are missing for almost all matches (probably those statistics were added in this season). I will drop them and check what is going on with other variables where are missings . Red and yellow cards are missing since if there weren't any cards in the match, the statistics were omitted on the page so I will fill it with 0.

```python
df = df.drop(columns=
                     ['Acc_pass_away','Acc_pass_home','Blocks_away','Blocks_home','Pass_away',
                          'Pass_home','Goal_kicks_away','Goal_kicks_home'])
df[['Red_away','Red_home','Yellow_away','Yellow_home']] = df[['Red_away','Red_home','Yellow_away','Yellow_home']].fillna(0)
```

# Adding result of the match column. Result is related to Home team.

```python
df['Result'] = np.where(df['Goals_home'] > df['Goals_away'], 'W', 
                           np.where(df['Goals_home'] == df['Goals_away'], 'D',
                               'L'))
print(df['Result'].unique())
```

# Check other variables with missing values. I will check only variables with 'home' suffix as missing values for variables with 'away' suffix are the same.

```python
df_nulls_other = df.groupby('Season')[['Pass_target_home','Dang_att_home','Attacks_home','Auts_home','Blocked_shots_home','FK_home','Fauls_home','Offsides_home','Ball_possession_home']].apply(lambda x : x.isnull().sum())
seasons_sizes = df.groupby('Season').size()
df_nulls_other = df_nulls_other.join(seasons_sizes.rename('n_matches'))
nulls_pct = df_nulls_other.iloc[:,:-1].div(df_nulls_other['n_matches'], axis = 0)
fig, ax = plt.subplots(nrows=9, ncols= 1, figsize=(100,250))

fig.tight_layout(pad=40.0)

i = 0
for col in nulls_pct.columns:
    ax[i].tick_params(axis='x', labelsize=70)
    ax[i].tick_params(axis='y', labelsize=80, labelrotation = 45)
    ax[i].set_ylim([0,1.0])
    ax[i].set_title(col, size = 80)
    nulls_pct[col].plot(kind='bar', ax=ax[i])
    i += 1
```

Most of those variables are missing for a few seasons so I will drop them. For "Fauls", 'Offsides' and 'Ball_possession' variables I will fill Null values later.

```python
df = df.drop(columns=['Pass_target_home','Pass_target_away',
                     'Dang_att_home', 'Dang_att_away',
                     'Attacks_home','Attacks_away',
                     'Auts_home', 'Auts_away',
                     'Blocked_shots_home', 'Blocked_shots_away',
                     'FK_home','FK_away'])
```

Check values in ball possession variable and convert it to numeric values

```python
print(df['Ball_possession_home'].unique())
```

```python
df['Ball_possession_home'] = df['Ball_possession_home'].str[:2]
df['Ball_possession_away'] = df['Ball_possession_away'].str[:2]
df[['Ball_possession_home','Ball_possession_away']] = df[['Ball_possession_home','Ball_possession_away']].astype(float)
```

```python
print(df[['Ball_possession_home','Ball_possession_away']])
```

Check matches where number of fauls or number of offsides is missing

```python
pd.set_option('display.max_columns', 50)
print(df[df['Fauls_home'].isnull()])
print(df[df['Offsides_home'].isnull()])
```

There are only 7 mathces wtih missing fauls variable and 15 (some of them are the same like with missing fauls) with missing offsides so I will fill try to add it manually (using data found on other page with statistics from matches). Let's also look at matches where ball possession is missing.

```python
df.loc[244,['Offsides_home','Offsides_away','Fauls_home','Fauls_away']] = [2,1,15,19]
df.loc[267,['Offsides_home','Offsides_away','Fauls_home','Fauls_away']] = [3,2,23,11]
print(df[df['Ball_possession_home'].isnull()])
```

That are quite old matches and unfortunatelly I couldn't find statistics for other matches. I will fill missing values with average value for each variable of the team in the season.

```python
df['Fauls_home'] = df.groupby(['Season','Team_Home']).Fauls_home.transform(lambda x: x.fillna(x.mean()))
df['Fauls_away'] = df.groupby(['Season','Team_Away']).Fauls_away.transform(lambda x: x.fillna(x.mean()))
df['Offsides_home'] = df.groupby(['Season','Team_Home']).Offsides_home.transform(lambda x: x.fillna(x.mean()))
df['Offsides_away'] = df.groupby(['Season','Team_Away']).Offsides_away.transform(lambda x: x.fillna(x.mean()))
df['Ball_possession_home'] = df.groupby(['Season','Team_Home']).Ball_possession_home.transform(lambda x: x.fillna(x.mean()))
df['Ball_possession_away'] = df.groupby(['Season','Team_Away']).Ball_possession_away.transform(lambda x: x.fillna(x.mean()))
print(df[['Ball_possession_home','Ball_possession_away']])
print(df[['Fauls_home','Fauls_away']])
```

# Change type of variables

```python
df_floats_list = list(df.select_dtypes(include='float64'))
df[df_floats_list] = df[df_floats_list].astype(int)
df[['Season','Team_Home','Team_Away', 'Result']] = df[['Season','Team_Home','Team_Away', 'Result']].astype('string')
df['Date'] = pd.to_datetime(df['Date'])
```
